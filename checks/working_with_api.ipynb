```python
from mcs import run_pack
from mcs.provenance import ArtifactRef, MetricSummary
```

```python
out = run_pack(
    "../examples/",
    train_file="train_classification.yaml",
    execution_file="execution_gpu.yaml",
    strict=False,
    artifact_mode="reference",
    artifacts=[
        ArtifactRef(kind="metrics", path="artifacts/metrics/cls.json", format="json"),
        ArtifactRef(kind="predictions", path="artifacts/predictions/cls.parquet", format="parquet"),
    ],
    metric_summary=MetricSummary(split="test", metrics={"roc_auc": 0.91}),
    notes="Case study: classification baseline.",
)

print(out["record"].run_id)
print(out["registry_paths"]["run"])
```
